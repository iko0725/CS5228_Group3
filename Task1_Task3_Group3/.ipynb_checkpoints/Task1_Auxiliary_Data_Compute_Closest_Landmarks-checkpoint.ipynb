```python
# Compute distance based on lat and lng
def compute_distance(lat1, lng1, lat2, lng2):
    return(geopy.distance.geodesic((lat1, lng1), (lat2, lng2)).km)
```

```python
def compute_distance_ith(lat, lng, df, i):
    return np.vectorize(compute_distance)(lat, lng, df['lat'][i], df['lng'][i])
```

```python
def compute_auxiliary(DATA, train_df, test_df):
    # Set data, train or test
    if DATA == 'train':
        lat = train_df.lat
        lng = train_df.lng
    else:
        test_df['lat'].mask(test_df.lat>3, other = 1.31658333333333, inplace=True)
        test_df['lng'].mask((test_df.lng>104)|(test_df.lng<102), other = 103.85602777777778, inplace=True)
        lat = test_df.lat
        lng = test_df.lng
    
    # initialization
    max_distance = 10000
    num_records = len(lat)
    dist_comercial = np.zeros(num_records) + max_distance
    dist_primary = np.zeros(num_records) + max_distance
    dist_secondary = np.zeros(num_records) + max_distance
    dist_mall = np.zeros(num_records) + max_distance
    dist_mrt = np.zeros(num_records) + max_distance

    for i in range(len(commercial_centres)):
        d = compute_distance_ith(lat, lng, commercial_centres, i)
        dist_comercial[(d < dist_comercial)] = d[(d < dist_comercial)]
    
    for i in range(len(primary_school)):
        d = compute_distance_ith(lat, lng, primary_school, i)
        dist_primary[(d < dist_primary)] = d[(d < dist_primary)]
    
    for i in range(len(secondary_school)):
        d = compute_distance_ith(lat, lng, secondary_school, i)
        dist_secondary[(d < dist_secondary)] = d[(d < dist_secondary)]
        
    for i in range(len(shopping_malls)):
        d = compute_distance_ith(lat, lng, shopping_malls, i)
        dist_mall[(d < dist_mall)] = d[(d < dist_mall)]

    for i in range(len(mrt_station)):
        d = compute_distance_ith(lat, lng, mrt_station, i)
        dist_mrt[(d < dist_mrt)] = d[(d < dist_mrt)]

    dist_comercial = dist_comercial.reshape(num_records,1)
    dist_primary = dist_primary.reshape(num_records,1)
    dist_secondary = dist_secondary.reshape(num_records,1)
    dist_mall = dist_mall.reshape(num_records,1)
    dist_mrt = dist_mrt.reshape(num_records,1)
        
    concatenate_array = np.concatenate((dist_cc,dist_ps,dist_ss,dist_sm,dist_ts),axis=1)
    closest_distance = pd.DataFrame(concatenate_array, columns = ['closest_comercial','closest_primary','closest_secondary','closest_mall','closest_mrt'])
    return closest_distance
```

## Compute Centrality in Transportation Networks

```python
import re

mrt = pd.read_csv('./data/auxiliary-data/train-station.csv')

mrt.drop(columns=['Unnamed: 0'],axis=1, inplace=True)
mrt.loc[:,'line_code'] = mrt.loc[:,'stn_code'].apply(lambda x : "".join(re.findall("[a-zA-Z]+", x)))
mrt.loc[:,'mrt_station'] = mrt.loc[:,'mrt_station'].apply(lambda x : str(x).lower())
mrt_station.loc[:,'name'] = mrt_station.loc[:,'name'].apply(lambda x : str(x).lower())
```

```python
line_list = mrt['line_code'].unique()
```

```python
mrt
```

```python
mrt_station
```

```python
singapore_mrt_station = pd.merge(mrt, mrt_station, left_on='mrt_station', right_on='name')
singapore_mrt_station = singapore_mrt_station[['stn_code','line_code',  'name', 'lat', 'lng']]
singapore_mrt_station
```

```python
# Build Graph
import networkx as nx
G = nx.DiGraph()
mrt_names = list(singapore_mrt_station['name'].unique())
G.add_nodes_from(mrt_names)
```

```python
singapore_mrt_station
```

```python
singapore_mrt_station.duplicated().sum()
```

```python
singapore_mrt_station = singapore_mrt_station.drop_duplicates()
singapore_mrt_station
```

```python
import re
def count_station_number(txt):
    l = re.findall(r'\d+', txt)
    if txt == 'STC' or txt == 'PTC':
        return 0
    else: 
        return int(l[0])

for line in line_list:
    line_mrt = singapore_mrt_station[singapore_mrt_station['line_code'] == line]
    line_mrt.loc[:, 'stn_code'] = line_mrt.loc[:, 'stn_code'].apply(count_station_number)
    line_mrt = line_mrt.set_index('stn_code').sort_index()
    for i in range(len(line_mrt.index)-1):
        dist = compute_distance(line_mrt.at[line_mrt.index[i], 'lat'], 
                                line_mrt.at[line_mrt.index[i], 'lng'], 
                                line_mrt.at[line_mrt.index[i+1], 'lat'], 
                                line_mrt.at[line_mrt.index[i+1], 'lng'])
        node1 = line_mrt.at[[line_mrt.index[i]][0], 'name']
        node2 = line_mrt.at[[line_mrt.index[i+1]][0], 'name']
        G.add_edge(node1, node2, weight = dist)
        G.add_edge(node2, node1, weight = dist)
```

```python
# convert MRT maps to graphsb
plt.figure(figsize=(30, 12))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, cmap=plt.get_cmap('jet'), node_size = 500)
nx.draw_networkx_labels(G, pos)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), arrows=False)
plt.show()
```

```python
import operator
```

```python
closeness = sorted(nx.centrality.closeness_centrality(G).items(), key=operator.itemgetter(1),reverse=True)
```

```python
betweenness = sorted(nx.centrality.betweenness_centrality(G).items(), key=operator.itemgetter(1),reverse=True)
```

```python
closeness = {x[0]:x[1] for x in closeness}
```

```python
dist_ts_c = np.zeros(len(lat))
dist_ts_b = np.zeros(len(lat))
dist_ts = np.zeros(len(lat)) + 10000

for i in range(len(mrt_station)):
        d = compute_distance_ith(lat, lng, mrt_station, i)
        current_stn = list(mrt_station['name'])[i]
        if current_stn in closeness.keys():
            choose_station_idx = (d < dist_ts)
            dist_ts[choose_station_idx] = d[choose_station_idx]
            temp = 1/dist_ts
            dist_ts_c[choose_station_idx] = temp[choose_station_idx] * closeness[current_stn]
            dist_ts_b[choose_station_idx] = temp[choose_station_idx] * betweenness[current_stn]
            
d = {'closeness_closest_mrt': dist_ts_c, 'betweenness_closest_mrt': dist_ts_b}
weighted_closeness_amenties = pd.DataFrame(data=d)
```
