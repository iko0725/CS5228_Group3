## Step 1: Count the Number of Nearby landmarks

```python
import numpy as np
import pandas as pd
import geopy
from geopy.distance import geodesic
```

```python
# from google.colab import drive
# import os
# drive.mount('/content/drive')
# os.chdir('/content/drive/MyDrive/Colab/CS5228')
```

```python
#auxiliary-data
commerical = pd.read_csv('./data/auxiliary-data/sg-commerical-centres.csv')
mrt = pd.read_csv('./data/auxiliary-data/sg-mrt-stations.csv')
primary = pd.read_csv('./data/auxiliary-data/sg-primary-schools.csv')
secondary = pd.read_csv('./data/auxiliary-data/sg-secondary-schools.csv')
shopping = pd.read_csv('./data/auxiliary-data/sg-shopping-malls.csv')
```

```python
def count_num(data,auxiliary):
    num = 0
    for index,row in auxiliary.iterrows():
        distance = geodesic((data['lat'],data['lng']),(row['lat'],row['lng']))
        #Radio:1 km
        if distance<1:
            num = num+1
    return num
```

For train data

```python
#train_data
train_data = pd.read_csv('train.csv')
train_data.head()
```

```python
num = 0
for index,station in mrt.iterrows():
    distance = geodesic((train_data['lat'][28],train_data['lng'][28]),(station['lat'],station['lng']))
    if distance<1:
        num = num+1
        print(distance)
        print(station['name'])
    
num
```

```python
count_num(train_data.iloc[280,:],shopping)
```

```python
count_mrt_station = train_data.apply(lambda x:count_num(x,mrt),axis=1)
count_mrt_station.to_csv('count_mrt.csv')
```

```python
count_shopping = train_data.apply(lambda x:count_num(x,shopping),axis=1)
count_shopping.to_csv('count_shopping.csv')
```

```python
count_commerical = train_data.apply(lambda x:count_num(x,commerical),axis=1)
count_commerical.to_csv('count_commerical.csv')
```

```python
count_primary = train_data.apply(lambda x:count_num(x,primary),axis=1)
count_primary.to_csv('count_primary.csv')
```

```python
count_secondary = train_data.apply(lambda x:count_num(x,secondary),axis=1)
count_secondary.to_csv('count_secondary.csv')
```

```python
id = train_data['listing_id']
```

```python
auxiliary_count = pd.concat([id,count_commerical,count_mrt_station,count_primary,count_secondary,count_shopping],axis=1)
auxiliary_count.columns = ['listing_id','count_commerical','count_mrt_station','count_primary','count_secondary','count_shopping']
auxiliary_count.to_csv('auxiliary_count_train.csv',index=None)
auxiliary_count.head()
```
