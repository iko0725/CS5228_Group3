```python
import numpy as np
import pandas as pd

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor

import matplotlib.pyplot as plt

from sklearn import tree
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split
```

```python
def data_static(df,plottype = 'hist'):
    if plottype == 'hist':
        fig1, axs = plt.subplots(2, 4, sharey=True, tight_layout=True)
        axs[0,0].hist(df[:,0], bins=10)
        axs[0,0].set_title('Sales')
        axs[0,1].hist(df[:,1], bins=15)
        axs[0,1].set_title('CompPrice')
        axs[0,2].hist(df[:,2], bins=15)
        axs[0,2].set_title('Income')
        axs[0,3].hist(df[:,3], bins=15)
        axs[0,3].set_title('Advertising')
        axs[1,0].hist(df[:,4], bins=15)
        axs[1,0].set_title('Population')
        axs[1,1].hist(df[:,5], bins=15)
        axs[1,1].set_title('Price')
        axs[1,2].hist(df[:,7], bins=15)
        axs[1,2].set_title('Age')
        axs[1,3].hist(df[:,8], bins=8)
        axs[1,3].set_title('Education')
        plt.show()

    elif plottype == 'box':
        fig1, axs = plt.subplots(2, 4, sharey=True, tight_layout=True)
        axs[0,0].boxplot(df[:,0])
        axs[0,0].set_title('Sales')
        axs[0,1].boxplot(df[:,1])
        axs[0,1].set_title('CompPrice')
        axs[0,2].boxplot(df[:,2])
        axs[0,2].set_title('Income')
        axs[0,3].boxplot(df[:,3])
        axs[0,3].set_title('Advertising')
        axs[1,0].boxplot(df[:,4])
        axs[1,0].set_title('Population')
        axs[1,1].boxplot(df[:,5])
        axs[1,1].set_title('Price')
        axs[1,2].boxplot(df[:,7])
        axs[1,2].set_title('Age')
        axs[1,3].boxplot(df[:,8])
        axs[1,3].set_title('Education')
        plt.show()

    elif plottype == 'threefeature_bar':
        df_6 = list()
        df_9 = list()
        df_10 = list()
        df_6.append((np.sum(df[:,6] == 0)))
        df_6.append((np.sum(df[:,6] == 1)))
        df_6.append((np.sum(df[:,6] == 2)))
        df_9.append((np.sum(df[:,9] == 0)))
        df_9.append((np.sum(df[:,9] == 1)))
        df_10.append((np.sum(df[:,10] == 0)))
        df_10.append((np.sum(df[:,10] == 1)))

        plt.bar(x=0, height= df_6[0], width=0.45, alpha=0.8, color='red', label="0")
        plt.bar(x=0, height=df_6[1], width=0.45, alpha=0.8, color='blue', label="1",bottom = df_6[0])
        plt.bar(x=0, height=df_6[2], width=0.45, alpha=0.8, color='darkred', label="2",bottom = df_6[1]+df_6[0])
        plt.bar(x=1, height= df_9[0], width=0.45, alpha=0.8, color='red')
        plt.bar(x=1, height=df_9[1], width=0.45, alpha=0.8, color='blue', bottom = df_9[0])

        plt.bar(x=2, height= df_10[0], width=0.45, alpha=0.8, color='red')
        plt.bar(x=2, height=df_10[1], width=0.45, alpha=0.8, color='blue',bottom = df_10[0])

        plt.xticks([0,1,2], ['ShelveLoc','Urban','US'])
        plt.legend()
        plt.show()

    elif plottype == 'scatter':
        import seaborn as sns
        col_name = ["num_beds","num_baths","size_sqft","price"]
        sns.pairplot(pd.DataFrame(df, columns = col_name[0:4]), vars=col_name[0:4])  
        plt.show()

    elif plottype == 'norm':
        from scipy.stats import norm
        feature_index = [0,1,5]
        fig1, axs = plt.subplots(1, 3, sharey=True, tight_layout=True)
        for i in [0,1,2]:
            x = df[:,feature_index[i]].tolist()
            n, bins, patches = axs[i].hist(x, 10,facecolor='blue', alpha=0.5,density=True) 
            axs[i].plot(bins,norm.pdf(bins, np.mean(x), np.std(x)), 'r--',label = 'Normal Distribution') 
        axs[0].set_title('Sales')
        axs[1].set_title('CompPrice')
        axs[2].set_title('Price')
        plt.legend()
        plt.show()

```
