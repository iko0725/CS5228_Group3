### User Based CF

Randomly produce user profile.
With both positive ratings and negative ratings.

```python
df_track2 = df_sample.copy()
df_track2['per_price'] = df_track2['price']/df_track2['size_sqft']
np.percentile(df_track2['per_price'],[25,50,75])
```

Assume that user prefer `built_year` after 2000 and `per_price` lower than 2500(As 2506 is the 75% quantile of `per_price`).

```python
np.random.seed(123)
#randomly generate 500 users' data

n = 500
#positive rating for select houses, for houses with per_price<=2500 & built_year>=2000.
user_select = df_track2[(df_track2.per_price<=2500)&(df_track2.built_year>=2000)]
user_item = user_select.sample(n=n)
pos_score = np.arange(3.0, 5.5, 0.5)
user_item_index = user_item['listing_id'].tolist()
user_item['rating'] = np.random.choice(pos_score,n)
user_item['user_id'] = range(n)
```

Assume that user give low ratings for houses `built_year` after 2000 and no mrt station near within 1 km.

```python
#negative rating for select houses, for houses with count_mrt_station==0 & built_year<2000
user_select_neg = df_track2[(df_track2.count_mrt_station==0)&(df_track2.built_year<2000)]
user_item_neg = user_select_neg.sample(n=n)
neg_score = np.arange(0, 3.0, 0.5)
user_item_neg_index = user_item_neg['listing_id'].tolist()
user_item_neg['rating'] = np.random.choice(neg_score,n)
user_item_neg['user_id'] = range(n)
```

```python
#combine the two data frame with user_id and ratings.
combine_data = pd.concat([user_item,user_item_neg])
combine_data.sample(5)
```

```python
num_house = df_track2.shape[0]
user_profiles = np.zeros((n, num_house))
M = np.zeros((n, num_house))
M.shape
```

```python
#fill in with created ratings 
for index, row in combine_data.iterrows():
    user_id, house_id, rating = row['user_id'], row.name, row['rating']
    M[user_id][house_id] = rating

M[152][13377]
```

```python
masked = np.ma.masked_equal(M, 0)
house_mean_ratings = np.mean(masked, axis=0)
user_mean_ratings = np.mean(masked, axis=1).reshape(-1, 1)
M_normalized = M - user_mean_ratings
M_normalized[M == 0] = 0
user_similarities = cosine_similarity(M_normalized, M_normalized)
user_similarities.shape
```

## Computing the Top Recommendations

```python
def get_top_recommendations(row, **kwargs) -> pd.DataFrame:
    
    #####################################################
    ## Initialize the required parameters
    
    topk = kwargs.get('topk')
    factor = kwargs.get('factor')  

            
       
    #####################################################
    ## Compute your recommendations
    #

    user_idx = range(n)

    recommendations = np.argsort(user_similarities[user_idx].squeeze())
    recommendations = recommendations[::-1][:topk*factor]
    recommendations = recommendations[np.random.choice(range(recommendations.shape[0]), size=topk,replace=False)]
    recom_index = recommendations[-1][1:topk+1]
    
    df_result = df_sample.iloc[recom_index]
        
    # Return the dataset with the k recommendations
    return df_result
```

## Testing the Recommendation Engine

## Compute and Display the recommendations

## Computing the Top Recommendations

```python
def get_top_recommendations(row, **kwargs) -> pd.DataFrame:
    
    #####################################################
    ## Initialize the required parameters
    
    topk = kwargs.get('topk')
    df = kwargs.get('df')
    users = row.user_id
            
       
    #####################################################
    ## Compute your recommendations
    #

    df_copy = df.copy()
    recom_data = tc.SFrame(df_copy)
    model = tc.item_similarity_recommender.create(recom_data, user_id='user_id', item_id='listing_id', target='rating', similarity_type='cosine', verbose=False)
    listing_id = model.recommend(users=[users],k=topk)['listing_id']

    df_result = combine_data[combine_data['listing_id'].apply(lambda i: i in listing_id)]
        
    # Return the dataset with the k recommendations
    return df_result
```
